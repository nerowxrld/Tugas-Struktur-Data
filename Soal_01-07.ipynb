```python
import matplotlib.pyplot as plt
```

1. masukan 10 nilai

```python
nilai = []

for i in range(10):
    n = int(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
    nilai.append(n)
```

data nilai mahasiswa

```python
print("\nData Nilai Mahasiswa:", nilai)
```

2. nilai tertinggi dan terendah

```python
nilai_tertinggi = max(nilai)
nilai_terendah = min(nilai)

print("Nilai tertinggi:", nilai_tertinggi)
print("Nilai terendah:", nilai_terendah)
```

3. nilai rata rata

```python
rata_rata = sum(nilai) / len(nilai)
print("Rata-rata nilai:", rata_rata)
```

4. hitung mahasiswa yang lulus dan tidak lulus

```python
lulus = 0
tidak_lulus = 0

for n in nilai:
    if n >= 60:
        lulus += 1
    else:
        tidak_lulus += 1

print("Jumlah mahasiswa lulus:", lulus)
print("Jumlah mahasiswa tidak lulus:", tidak_lulus)
```

5. grafik gambar nilai tertinggi dan terendah

```python
label = ['Nilai Tertinggi', 'Nilai Terendah']
data = [nilai_tertinggi, nilai_terendah]

plt.figure()
plt.bar(label, data)
plt.title("Perbandingan Nilai Tertinggi dan Terendah")
plt.ylabel("Nilai")
plt.show()
```

6. grafik data kelulusan

```python
label_kelulusan = ['Lulus', 'Tidak Lulus']
data_kelulusan = [lulus, tidak_lulus]

plt.figure()
plt.pie(data_kelulusan, labels=label_kelulusan, autopct='%1.1f%%')
plt.title("Persentase Kelulusan Mahasiswa")
plt.show()
```
